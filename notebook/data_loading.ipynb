## Data Cleaning 

```python
## Droping NaN values of TotalCharges
df.dropna(subset=['TotalCharges'], inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
# Remove customerID column
df = df.drop('customerID', axis=1)
df.head()
```

CustomerID is a unique column which does not have any involvement in churn prediction so this column is removed.

```python
## Separating the target varibale 
y = df['Churn'].map({'No': 0, 'Yes': 1})
X = df.drop('Churn', axis=1)
```

### Converting Categorical Variables to Numerical

```python
# Identifying categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
print(categorical_cols)
```

Now these columns should be encoded numerically. For binary categorical varibale we use binary encoding and one hot encoding for other categorical varibale.  

```python
# Binary Encoding
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']

for col in binary_cols:
    X[col] = X[col].map({'Yes': 1, 'No':0})
```

```python
# Exclude the binary columns already encoded
multi_cat_cols = [col for col in categorical_cols if col not in binary_cols]

print("Columns to one-hot encode:", multi_cat_cols)
```

```python
# One Hot Encoding
X = pd.get_dummies(X, columns=multi_cat_cols, drop_first=True, dtype=int)
X.head()
```

```python
X.info()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

# Splitting data : 70% train, 30% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

```python
# Check the shape of splits
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

## Saving split dataset to files 

```python
import pickle

with open('X_train.pkl', 'wb') as f:
    pickle.dump(X_train, f)

with open('X_test.pkl', 'wb') as f:
    pickle.dump(X_test, f)

with open('y_train.pkl', 'wb') as f:
    pickle.dump(y_train, f)

with open('y_test.pkl', 'wb') as f:
    pickle.dump(y_test, f)
```
